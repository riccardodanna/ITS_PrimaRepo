```python
valori_x = []  # seconda versione: valori_x = [x for x in range(-20, 21)]
for x in range(-20,21):   # in altri linguaggi di programmazione(int x = -20: x < 21; x+=1)
    valori_x.append(x)

print(valori_x)
```

```python
import numpy as np

array_x = np.array(valori_x)
array_retta = 3 * array_x

print(array_x)
print(array_retta)
```

```python
# creazione grafico
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,6))
ax_retta = fig.add_axes(rect=(0,0,1,1)) # questa parentesi è un tuple: lista che usa le parentesi() invece di [] e non è modificabile
ax_retta.plot(array_x, array_retta)
```

```python
# esercizio: modificare il codice e aggiungere una parabola al grafico (x^2)
# aggiungere una parabola (x^2 + 2x + 6)

array_parabola = array_x**2

fig = plt.figure(figsize=(12,6))
ax_retta = fig.add_axes(rect=(0,0,1,1))
ax_retta.plot(array_x, array_retta)
ax_retta.plot(array_x, array_parabola)
```

```python
array_parabola2 = array_x**2 + array_x*2 + 6
ax_retta.plot(array_x, array_parabola2)
fig
```

```python
# alternativa per più grafici
fig, axes = plt.subplots(nrows=3, ncols=1)
axes[0].plot(array_x, array_retta, color = 'blue')
axes[1].plot(array_x, array_parabola, color = 'green')
axes[2].plot(array_x, array_parabola2, color = 'red')
```
