```python
import pandas as pd

df_flights = pd.read_csv('data/flights.csv')
df_flights
```

### Deal with null values

```python
df_flights.isnull().sum()

df_flights[df_flights.isnull().any(axis=1)]

df_flights.DepDel15 = df_flights.DepDel15.fillna(0.0)   # sostituisco i valori nulli con 0.0
```

```python
df_flights.isnull().sum()     # controllo se ci sono ancora valori nulli
```

### Deal with outliers

```python
# Find mean, median, mode, maximum and minimum for DepDelay and ArrDelay columns

dep_mean = df_flights.DepDelay.mean()
dep_median = df_flights.DepDelay.median()
dep_mode = df_flights.DepDelay.mode()
dep_max = df_flights.DepDelay.max()
dep_min = df_flights.DepDelay.min()

arr_mean = df_flights.ArrDelay.mean()
arr_median = df_flights.ArrDelay.median()
arr_mode = df_flights.ArrDelay.mode()
arr_max = df_flights.ArrDelay.max()
arr_min = df_flights.ArrDelay.min()

print('Departures data:')
print(dep_mean)
print(dep_median)
print(dep_mode)
print(dep_max)
print(dep_min)

print('\nArrivals data:')
print(arr_mean)
print(arr_median)
print(arr_mode)
print(arr_max)
print(arr_min)
```

```python
# plot distribuzione
import matplotlib.pyplot as plt

def show_distribution (col):
    delay_mean = col.mean()
    delay_median = col.median()
    delay_mode = col.mode()
    delay_max = col.max()
    delay_min = col.min()
    print(col.name,'\nMin:{}\nMax:{}\nMean:{}\nMedian:{}\nMode:{}'.format(delay_min, delay_max, delay_mean, delay_median, delay_mode))

    fig,ax = plt.subplots(1, 2, figsize=(12,6))

    ax[0].hist(col)
    ax[0].set_ylabel('Frequenza')

    ax[1].axvline(x = delay_min, color='grey', linestyle='dashed', Linewidth=2)
    ax[1].axvline(x = delay_max, color='green', linestyle='dashed', Linewidth=2)
    ax[1].axvline(x = delay_mean, color='red', linestyle='dashed', Linewidth=2)
    ax[1].axvline(x = delay_median, color='blue', linestyle='dashed', Linewidth=2)
    ax[1].axvline(x = delay_mode, color='yellow', linestyle='dashed', Linewidth=2)

    
```
